## DATA PREPROCESSING STEPS

```python
# Import Libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Import Dataset

dataset=pd.read_csv("The_Grades_Dataset.csv")
```

```python
dataset #display dataset
```

```python
#Dropping Unnecessary column
# Droping Seat No. Column

df=dataset.drop(['Seat No.'],axis="columns")
```

```python
df #display dataset
```

```python
# Checking for the missing values in the dataset

df.info()
```

```python
#PLOTTING NULL VALUES
plt.figure(figsize=(25,25))
sns.heatmap(dataset.isnull())
```

```python
# checking for total missing values count at each row in DataFrame

for i in range(len(df.index)) :
        print("NaN in row ", i , " : " ,  df.iloc[i].isnull().sum())
```

```python
#droping those rows that have more than 10 missing values

df.drop([44,60,91,137,142,143,281,282,288,565], axis=0, inplace=True)
```

```python
# checking again for total missing values count at each row in DataFrame

for i in range(len(df.index)) :
        print("NaN in row ", i , " : " ,  df.iloc[i].isnull().sum())
```

```python
# Using Mode to fill the missing categorical values

for i in range(41):
    this_column = df.columns[i]
    df[this_column] = df[this_column].fillna(df[this_column].mode()[0])
df1=df
df1
```

```python
#PLOTTING NULL VALUES
plt.figure(figsize=(25,25))
sns.heatmap(df1.isnull())
```

```python
#Using mapping technique to convert ordinal data into numerical values

dataMapping = {"A+":4.0, "A":4.0, "A-":3.7, "B+":3.4, "B":3.0, "B-":2.7, "C+":2.4, "C":2.0, "C-":1.7, "D+":1.4, "D":1.0, "F":0.0, "I":0.0, "W":0.0, "WU":0.0}
for i in range(41):
    this_column = df1.columns[i]
    df1[this_column]=df1[this_column].map(dataMapping)
df1
```

```python
# SEPERATING FEATURE VALUES(X) AND TARGET VALUES(Y)

X=df.iloc[:, :-1].values
Y=df.iloc[:,41].values
```

```python
# SPLITTING DATASET IN TRAINING SET AND TESTING SET

from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.30, random_state=0)
```

```python
# FEATURE SCALING USING STANDARD SCALER

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

### IMPLEMENTING MULTILPLE LINEAR REGRESSION TO CHECK ACCURACY

```python
# applying multiple linear regression algorithm to check for accuracy of the above modified dataset

from sklearn.linear_model import LinearRegression
r=LinearRegression()
r.fit(X_train,Y_train)
r.fit(X_test,Y_test)
print('The accuracy after training model on X_train and Y_train set:',r.score(X_train,Y_train))
print('The accuracy after training model on X_train and Y_train set:',r.score(X_test,Y_test))
```

```python
y_pred=r.predict(X_test)     # predicting test set values
from sklearn.metrics import r2_score   #import R2_score metric
r2 = r2_score(Y_test,y_pred)    #checking accuracy of predicted values
print(r2)
```

```python
plt.scatter(Y_test,y_pred)    # plotting predicted vs actual values
plt.title("Plot of predicted vs actual values")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
```
